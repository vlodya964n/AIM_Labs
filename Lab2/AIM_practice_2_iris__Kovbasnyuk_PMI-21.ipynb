```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

def train_model(X_train, X_test, y_train, y_test):

    # Ñ„ÑƒÐ½ÐºÑ†Ð¸Ñ Ð¾Ð±ÑƒÑ‡Ð°ÐµÑ‚ Ð¼Ð¾Ð´ÐµÐ»ÑŒ Ð¿Ð¾ Ð¿Ñ€Ð´Ð¾ÑÑ‚Ð°Ð²Ð»ÐµÐ½Ð½Ñ‹Ð¼ Ð²Ñ‹Ð±Ð¾Ñ€ÐºÐ°Ð¼ Ð¸ Ð²Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÑ‚ Ð¾Ñ†ÐµÐ½ÐºÑƒ Ñ‚Ð¾Ñ‡Ð½Ð¾ÑÑ‚Ð¸

    classifier = KNeighborsClassifier(n_neighbors=1)

    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    accuracy = round(accuracy * 100, 1)
    return accuracy

def use_preproc(preproc, preproc_name):

    # Ñ„ÑƒÐ½ÐºÑ†Ð¸Ñ Ð¿Ñ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ñ‹Ð²Ð°ÐµÑ‚ Ð½Ð°Ñ‡Ð°Ð»ÑŒÐ½Ñ‹Ðµ Ð´Ð°Ð½Ð½Ñ‹Ðµ

    X_train, X_test, y_train, y_test = train_test_split(iris_data, iris.target, test_size = 0.5, random_state = 0)

    X_train = preproc.fit_transform(X_train)

    X_test = preproc.transform(X_test)

    accuracy = train_model(X_train, X_test, y_train, y_test)

    results["preprocessing"].append(preproc_name)
    results["accuracy, %"].append(accuracy)
```

```python
results = {
    "preprocessing" : [],
    "accuracy, %" : []
}
```

```python

from sklearn.preprocessing import MaxAbsScaler

use_preproc(MaxAbsScaler(), "MaxAbsScaler")
```

```python
from sklearn.preprocessing import StandardScaler

use_preproc(StandardScaler(), "StandardScaler")
```

```python
from sklearn.preprocessing import QuantileTransformer

use_preproc(QuantileTransformer(random_state=0), "QuantileTransformer")

use_preproc(QuantileTransformer(output_distribution='normal', random_state=0), "normal distribution with QuantileTransformer")
```
