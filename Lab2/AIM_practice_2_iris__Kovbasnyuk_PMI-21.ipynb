#### Преобразование данных и обучение модели

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

def train_model(X_train, X_test, y_train, y_test):

    # функция обучает модель по прдоставленным выборкам и вычисляет оценку точности

    classifier = KNeighborsClassifier(n_neighbors=1)

    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    accuracy = round(accuracy * 100, 1)
    return accuracy

def use_preproc(preproc, preproc_name):

    # функция преобразовывает начальные данные

    X_train, X_test, y_train, y_test = train_test_split(iris_data, iris.target, test_size = 0.5, random_state = 0)

    X_train = preproc.fit_transform(X_train)

    X_test = preproc.transform(X_test)

    accuracy = train_model(X_train, X_test, y_train, y_test)

    results["preprocessing"].append(preproc_name)
    results["accuracy, %"].append(accuracy)
```

```python
results = {
    "preprocessing" : [],
    "accuracy, %" : []
}
```

#### Применение разных преобразований данных

```python

from sklearn.preprocessing import MaxAbsScaler

use_preproc(MaxAbsScaler(), "MaxAbsScaler")
```

```python
from sklearn.preprocessing import StandardScaler

use_preproc(StandardScaler(), "StandardScaler")
```

```python
from sklearn.preprocessing import MinMaxScaler

use_preproc(MinMaxScaler(), "MinMaxScaler")
```

```python
from sklearn.preprocessing import RobustScaler

use_preproc(RobustScaler(), "RobustScaler")
```

```python
from sklearn.preprocessing import QuantileTransformer

use_preproc(QuantileTransformer(random_state=0), "QuantileTransformer")

use_preproc(QuantileTransformer(output_distribution='normal', random_state=0), "normal distribution with QuantileTransformer")
```

```python
from sklearn.preprocessing import PowerTransformer

use_preproc(PowerTransformer(), "PowerTransformer yeo-johnson")

use_preproc(PowerTransformer(method='box-cox'), "PowerTransformer box-cox")
```

```python
from sklearn.preprocessing import Normalizer

use_preproc(Normalizer(), "Normalizer")
```

```python
from sklearn.preprocessing import KBinsDiscretizer

use_preproc(KBinsDiscretizer(encode='ordinal'), "quantile strategy KBinsDiscretizer")

use_preproc(KBinsDiscretizer(encode='ordinal', strategy='uniform'), "uniform startegy KBinsDiscretizer")

use_preproc(KBinsDiscretizer(encode='ordinal', strategy='kmeans'), "kmeans startegy KBinsDiscretizer")
```

```python
from sklearn.preprocessing import Binarizer

use_preproc(Binarizer(), "Binarizer")
```

```python
from sklearn.preprocessing import PolynomialFeatures

use_preproc(PolynomialFeatures(), "PolynomialFeatures")
```

```python
from sklearn.preprocessing import SplineTransformer

use_preproc(SplineTransformer(), "SplineTransformer")
```
